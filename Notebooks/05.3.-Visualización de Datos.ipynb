
El método `.plot` en Pandas es una funcionalidad que permite generar visualizaciones directamente desde un DataFrame o una Serie. Utiliza la librería Matplotlib internamente para crear gráficos básicos, lo que ofrece una forma rápida y sencilla de representar datos visualmente.

### Pros del Método `.plot`:

- **Simplicidad:** Es fácil de usar y no requiere un conocimiento profundo de Matplotlib.
- **Integración con Pandas:** Se puede aplicar directamente a DataFrames y Series.
- **Variedad de Gráficos:** Ofrece varios tipos de gráficos básicos, como líneas, barras, histogramas y más.
- **Personalización Básica:** Permite ajustar etiquetas, títulos y colores de manera sencilla.

### Contras del Método `.plot`:

- **Limitación en la Personalización:** No es tan versátil como Matplotlib para personalización avanzada.
- **Complejidad Limitada:** Más adecuado para gráficos simples; carece de opciones para visualizaciones complejas.
- **Rendimiento Limitado:** No es óptimo para conjuntos de datos grandes.
- **Limitaciones en Tipos de Gráficos:** Algunos tipos de gráficos más especializados no están disponibles.

En resumen, el método `.plot` de Pandas es útil para generar visualizaciones rápidas y sencillas, pero para visualizaciones más avanzadas y personalizadas, Matplotlib u otras librerías pueden ser más adecuadas.

# Diagrama de dispersión (pairplot)

<div class="alert alert-info" role="alert">
<b>Diagrama de dispersión</b>


Un **diagrama de dispersión**, también conocido como **gráfico de dispersión**, es una herramienta visual fundamental en el análisis de datos que utiliza las coordenadas cartesianas para representar los valores de dos variables en un conjunto de datos. Cada punto en el gráfico de dispersión corresponde a una observación o un par de valores de las dos variables, lo que permite visualizar la relación entre ellas.

### Principales Características:

- **Representación Bidimensional:** A diferencia de otros gráficos que muestran datos en una sola dimensión, el diagrama de dispersión presenta datos en dos dimensiones. El eje x (horizontal) representa una variable, mientras que el eje y (vertical) representa la otra variable.

- **Exploración de Correlación:** El principal propósito del diagrama de dispersión es ayudar a identificar patrones o relaciones entre las dos variables. Puede sugerir diferentes tipos de correlación, como:
  - **Correlación Positiva:** Cuando los puntos tienden a agruparse en una dirección ascendente en el gráfico, indicando que a medida que una variable aumenta, la otra también lo hace.
  - **Correlación Negativa:** Cuando los puntos tienden a agruparse en una dirección descendente, lo que indica que a medida que una variable aumenta, la otra disminuye.
  - **Correlación Nula:** Cuando los puntos no siguen un patrón claro y están dispersos al azar, lo que sugiere que las variables no están correlacionadas.

- **Identificación de Outliers:** El diagrama de dispersión también es útil para detectar valores atípicos (outliers) que se alejan significativamente del patrón general de dispersión de los puntos.

<div>

<b>Gráfico circular</b>



Un **gráfico circular**, ampliamente conocido como **gráfico de pastel**, es una valiosa herramienta de visualización estadística utilizada para representar porcentajes y proporciones en un conjunto de datos. Su nombre proviene de su característica forma circular, que se asemeja a un pastel o tarta dividida en rebanadas. Este tipo de gráfico es altamente efectivo en la comunicación de la distribución relativa de diversas categorías o componentes dentro de un conjunto de datos más amplio.

#### Principales Características:

- **Representación de Porcentajes y Proporciones:** El gráfico circular es especialmente eficaz para mostrar cómo se dividen y distribuyen las partes en relación con el todo. Cada segmento o rebanada del círculo representa una categoría o componente, y el tamaño angular de cada segmento es proporcional al porcentaje que representa en el conjunto de datos.

- **Facilidad de Comprensión:** Gracias a su diseño intuitivo y visualmente atractivo, el gráfico circular es fácil de comprender para una amplia audiencia. Facilita la comunicación de la distribución relativa de categorías de manera clara y concisa.

- **Versatilidad:** Puede utilizarse para representar una amplia gama de datos, desde la distribución de ingresos en un país hasta la composición de un presupuesto familiar. Es una herramienta versátil para comunicar la estructura y la proporción de cualquier conjunto de datos.

- **Múltiples Categorías:** Aunque se asocia comúnmente con cuatro o más categorías, el gráfico circular puede adaptarse para mostrar incluso pequeñas divisiones en datos complejos.


El gráfico circular más temprano conocido se atribuye generalmente al escocés William Playfair, en la obra Statistical Breviary de 1801.1​2​
    
</div>

`````python
from seaborn import color_palette  # Importar la función color_palette desde Seaborn

# Crear una figura con propiedades personalizadas
fig = figure(num=None, figsize=(8, 6), dpi=300, facecolor='w', edgecolor='k')

# Habilitar la cuadrícula en el gráfico
grid(True)

# Definir etiquetas y colores para las porciones del gráfico circular
labels = ['No Sobrevivientes', 'Sobrevivientes']

explode = linspace(0, 0.01, 2)  # Controla la explosión de las porciones

colors = color_palette('pastel')[0:2]  # Paleta de colores personalizada


# Contar las ocurrencias de sobrevivientes y no sobrevivientes
count_class = DataFrame(df.groupby(['Survived'])['Survived'].count())

# Ordenar los valores en orden descendente
df_ = count_class['Survived'].sort_values(ascending=False)

# Crear un gráfico de tipo pie chart
df_.plot.pie(legend=False, labels=labels, subplots=True, autopct='%.2f%%', colors=colors, startangle=20, explode=explode)

# Ajustar el gráfico
ylabel('')  # Eliminar etiqueta del eje y
axis('equal')  # Aspecto igualitario para un círculo
title('Porcentaje de Sobrevivientes y No Sobrevivientes en el Titanic', weight='bold', fontsize=12)
`````

<div class="alert alert-info" role="alert">
<b>Gráfico de Cajas (Box Plot)</b>

Un **box plot**, también conocido como **gráfico de caja y bigotes**, es una herramienta de visualización estadística que se utiliza para representar la distribución y la variabilidad de un conjunto de datos numéricos. Este gráfico recibe su nombre debido a su característica forma que se asemeja a una caja con "bigotes" en los extremos.

Esta herramienta gráfica utilizada en estadísticas y análisis de datos para representar la distribución y la dispersión de un conjunto de datos. Este gráfico proporciona una visión detallada de la mediana, los cuartiles y la presencia de valores atípicos en los datos.
    

#### Principales Características:

- **Representación de Distribución:** El box plot representa la distribución de un conjunto de datos mediante una caja rectangular y dos líneas, o "bigotes", que se extienden desde la caja. La caja muestra el rango intercuartílico `(IQR)`
, que contiene el `50%` central de los datos. La línea central de la caja representa la mediana.

- **Identificación de Outliers:** Los valores atípicos (outliers) se representan como puntos individuales más allá de los extremos de los bigotes, lo que facilita su identificación.

- **Visualización de Dispersión:** Los bigotes indican la variabilidad fuera del rango intercuartílico. Esto ayuda a visualizar si los datos están ampliamente dispersos o altamente concentrados alrededor de la mediana.

    
    

#### El gráfico de cajas consta de varios elementos clave:

- **Caja Central**: La caja representa el rango intercuartil `(IQR)`p, que abarca desde el primer cuartil `(Q1)` hasta el tercer cuartil `(Q3)`. El `IQR` contiene el `50%` central de los datos y muestra la dispersión de esta parte de los datos.

- **Línea en el Interior de la Caja**: Esta línea representa la mediana `(Q2)`, que es el valor que divide el conjunto de datos en dos partes iguales.

- **Bigotes (Whiskers)**: Los bigotes se extienden desde los extremos de la caja hasta los valores más extremos que no se consideran valores atípicos. Generalmente, se calculan como __1.5__ veces el `IQR` desde los cuartiles. Cualquier valor que esté más allá de los bigotes se considera un valor atípico y se muestra como un punto individual.

Puntos Individuales: Los puntos fuera de los bigotes se representan como valores atípicos y se muestran como puntos individuales en el gráfico.


  
</div>

<div class="alert alert-block alert-danger" role="alert">
    
<b>Estandarización y Normalización</b>
    
Aunque el gráfico de cajas puede utilizarse con datos sin procesar, en ocasiones es beneficioso preprocesar las variables antes de crear el gráfico para obtener una representación más clara y significativa de los datos. La necesidad de preprocesar las variables se debe a varias razones:

1. **Diferentes Escalas:** Si las variables tienen diferentes escalas o unidades de medida, la comparación directa en un gráfico de cajas puede ser engañosa. Preprocesar las variables para que tengan una escala común facilita la comparación.

2. **Outliers:** Los valores atípicos (outliers) pueden distorsionar la representación de la distribución en un gráfico de cajas. Algunas técnicas de preprocesamiento pueden ayudar a manejar los outliers.

3. **Interpretación Significativa:** Algunas variables pueden ser difíciles de interpretar en su forma original. El preprocesamiento puede transformar las variables para que su significado sea más claro.


Existen dos técnicas comunes de preprocesamiento de variables: __estandarización__ y __normalización__.


### Diferencia entre Estandarización y Normalización

La principal diferencia entre estandarización y normalización radica en cómo se escala la variable. La estandarización se enfoca en lograr una `media de 0` y una `desviación estándar de 1`, lo que facilita la comparación en términos de distribución y distancia relativa. La normalización, por otro lado, escala los valores al rango `[0, 1]`, conservando la interpretación en términos de proporciones.

La elección entre estas técnicas depende de los objetivos del análisis y la interpretación de los datos. Ambas son valiosas en el preprocesamiento de variables antes de crear gráficos de cajas u otras visualizaciones.

Ahora si estandarizamos las varibales  con el método `standardize()`  y graficamos de nuevo obtemos el siguiente gráfico:

`````python
# Importa la biblioteca 'preprocessing' de scikit-learn para la escalación de características
from sklearn import preprocessing

# Suponiendo que 'df' es un DataFrame de pandas con datos

# # Selecciona solo las columnas numéricas en 'df' y almacénalas en la variable 'x'
x = df_numerico =  df.select_dtypes(include=['number']).values  # Esto devuelve una matriz de NumPy con las características seleccionadas

# Crea una instancia del objeto MinMaxScaler para escalar las características en el rango [0, 1]
min_max_scaler = preprocessing.MinMaxScaler()

# Escala las características en 'x' utilizando MinMaxScaler
x_scaled = min_max_scaler.fit_transform(x)

# Crea un nuevo DataFrame 'df_min_max' a partir de las características escaladas
df_min_max = DataFrame(x_scaled)

# Genera un diagrama de caja (boxplot) para visualizar la distribución de las características escaladas
ax = df_min_max.boxplot()
`````

`````python
from scipy.stats import zscore
df_z_score = df.select_dtypes(include=['number']).apply(zscore)
ax = boxplot(data=df_z_score)
`````

<b>Gráfico scatterplot 3D</b>


La función `scatterplot` en 3D de Seaborn se utiliza para crear gráficos de dispersión tridimensionales que permiten visualizar la relación entre tres variables numéricas. A continuación, se proporciona una descripción de esta función:

- **Uso Principal**: `scatterplot` en 3D se utiliza principalmente para visualizar y analizar la relación tridimensional entre tres variables numéricas, lo que permite identificar patrones, tendencias y correlaciones en los datos en un espacio tridimensional.

- **Representación de Datos**: Cada punto en el gráfico representa un conjunto de valores (x, y, z) correspondientes a las tres variables en estudio. Esto permite observar la dispersión de los datos en un espacio tridimensional y detectar posibles agrupamientos o dispersiones en el espacio.

- **Personalización**: Al igual que en los gráficos de dispersión en 2D, se pueden personalizar varios aspectos del gráfico, como el color, el tamaño y el estilo de los puntos, para resaltar patrones tridimensionales.


`````python
# Importa la función 'colorbar' desde la biblioteca 'matplotlib.pyplot'
from matplotlib.pyplot import colorbar

# Crear un scatterplot en 3D para visualizar la relación entre la edad (age), la tarifa (fare) y la clase de pasajero (class)
# Configurar el estilo de fondo de la trama en blanco
set(style="whitegrid")

# Crear una figura con un tamaño personalizado (10x8 pulgadas)
fig = figure(figsize=(10, 8))

# Agregar un subgráfico 3D a la figura
ax = fig.add_subplot(111, projection='3d')

# Configurar el ángulo de vista en 3D (35 grados de elevación y 45 grados de azimut)
ax.view_init(45, 45)

# Crear un gráfico de dispersión en 3D con datos de edad, tarifa y clase de pasajero
scatter = ax.scatter(
    xs=df["Age"],  # Eje X: Edad
    ys=df["Fare"],  # Eje Y: Tarifa
    zs=df["Pclass"].astype("category").cat.codes,  # Eje Z: Clase de Pasajero
    c=df["Pclass"].astype("category").cat.codes,  # Color en función de la Clase de Pasajero
    cmap="Set2",  # Colormap utilizada para el color
    s=60,  # Tamaño de los esferas
    alpha=.2  # Transparencia de los puntos
)

# Personalizar el gráfico
ax.set_title("Relación tridimensional entre Edad, Tarifa y Clase de Pasajero en el Titanic")
ax.set_xlabel("Edad")
ax.set_ylabel("Tarifa")
ax.set_zlabel("Clase de Pasajero")

# Configurar las marcas en el eje Z para mostrar las categorías de clase de pasajero
ax.set_zticks(ticks=range(len(df["Pclass"].unique())))
ax.set_zticklabels(df["Pclass"].unique())

# Mostrar una barra de color que corresponda a la Clase de Pasajero
colorbar(scatter, label="Clase de Pasajero")
`````

En este ejemplo, hemos utilizado la función scatterplot en 3D para visualizar la relación tridimensional entre la edad, la tarifa y la clase de pasajero de los pasajeros del Titanic. 

Cada punto en el gráfico representa un pasajero, y el color y la posición en el eje z se utilizan para indicar la clase de pasajero. Esto nos permite observar la dispersión de las edades y las tarifas en un espacio tridimensional, así como su relación con la clase de pasajero.

Importar algunas bibliotecas necesarias

`````python
from pandas import read_excel
from sparklines import sparklines
from pandas import Grouper

# Descargar los datos
df = read_excel('https://github.com/chris1610/pbpython/blob/master/data/2018_Sales_Total.xlsx?raw=true')
`````

`````python
# Agrupar
monthly_sales = df.groupby([Grouper(key='date', freq='M')])['ext price'].agg(['sum']).reset_index()
monthly_sales['pct_of_total'] = monthly_sales['sum'] / df['ext price'].sum()
monthly_sales
`````

`````python
# Use a format dictionary to control formatting per column
format_dict = {'sum':'${0:,.0f}', 'date': '{:%m-%Y}', 'pct_of_total': '{:.2%}'}


monthly_sales.style.format(format_dict).hide()
`````

## Spark line

Definimos una función para utilizar Sparkline

`````python
from numpy import histogram as histogram_

def sparkline_str(x):
    bins=histogram_(x)[0]
    sl = ''.join(sparklines(bins))
    return sl

sparkline_str.__name__ = "sparkline"
`````

Luego podemos llamar a esta función como una función de agregación estándar:

`````python
df.groupby('name')[['quantity', 'ext price']].agg(['mean', sparkline_str])
`````
